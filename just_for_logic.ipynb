```python
# ADMIN CELL

import random

def assign_secret_santa(idx, order, poss, assd_to, assd_from, to_from, n):
    """
    backtracking function
    idx current position in order
    """
    if idx == n: # all done
        return True
    
    person = order[idx]
    
    for candidate in poss[person]:
        if assd_from[candidate]:
            continue
        # constraint propagation check, will this assignment trap anyone, leave them with 0 options
        valid = True
        for other in range(n):
            if not assd_to[other] and other != person:
                available_count = 0
                for option in poss[other]:
                    # option is available if not yet assigned and not the candidate we're about to assign
                    if not assd_from[option] and option != candidate:
                        available_count += 1
                
                # if other would have no valid options, skip this candidate
                if available_count == 0:
                    valid = False
                    break
        
        if not valid:
            continue
        
        # assigning this one
        to_from[person] = candidate
        assd_to[person] = True
        assd_from[candidate] = True
        
        # worked, entire complete
        if assign_secret_santa(idx + 1, order, poss, assd_to, assd_from, to_from, n):
            return True
        
        # this didn't work, backtrack
        del to_from[person]
        assd_to[person] = False
        assd_from[candidate] = False
    
    # no valid candidate found
    return False

print("enter names of santas, enter blank when done.")
people = []
count = 0
names = {}
indices = {}
mistakes = 0
while True:
    santa = input().strip()
    if santa=="":
        break
    if santa in people:
        print("you already entered this name, I'm ignoring this.")
        mistakes+=1
        if mistakes>5:
            print("cut it out! I suggest you start over.")
        continue
    people.append(santa)
    names[santa]=count
    indices[count]=santa
    count+=1
    print(santa)

print("if there are any people you don't want to pair up, enter their names now." \
"same as before, enter blank when done.:")
count = 1
excluded_pairs = []
while True:
    print(f"excluded pair {count}:")
    a = input("person a: ").strip()
    if a=="":
        break
    b = input("person b: ").strip()
    if b=="":
        break
    excluded_pairs.append([a,b])
    print(a, b)
    count+=1

ppl = []
for i in people:
    ppl.append(names[i])
n = len(ppl)

poss = {}
for i in range(n):
    lst = ppl.copy()
    random.shuffle(lst)
    poss[i] = lst  # initially all possible
    poss[i].remove(i)
for i in excluded_pairs:
    poss[names[i[0]]].remove(names[i[1]])
    poss[names[i[1]]].remove(names[i[0]])
order = sorted(range(n), key=lambda i: len(poss[i])) # sort by who has fewest options

assd_to = [False]*n # has been assigned some person to gift to
assd_from = [False]*n # has been drawn by some other person to receive gift from
to_from = {}

if assign_secret_santa(0, order, poss, assd_to, assd_from, to_from, n):
    print("assignments completed.")
else:
    print("no valid assignment possible with these constraints.")
```
